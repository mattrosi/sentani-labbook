```python
import pandas as pd
import numpy as np
import pysentani as sti
```

```python
survey = pd.read_excel('../data-anonymous/sentani-merged-cleaned-anonymous-2014-11-29.xlsx')
```

```python
survey['access_type'] = sti.access_type(survey)
survey['electricity_monthly'] = sti.elec_expenditure_monthly(survey)
```

```python
survey = survey[survey['village_name'] == 'Asei']
```

```python
#expenditure data has a large range which is hard to interpret
subsurvey = survey[['access_type','PLN_expenditure_monthly',
                    'app_now/TV','app_now/lighting','app_now/radio',
                    'app_now/rice_cooker','app_now/fridge','app_now/fan']]
subsurvey.dropna()
```

```python
subsurvey.groupby('access_type').mean().transpose()
```

$$expenditure\ in\ rp \left(\frac{1USD}{12,000rp}\right) \left(\frac{1kWh}{0.08USD}\right)=\left(\frac{expenditure}{960}\right)kWh $$

```python
# monthly expenditure divided by 960 rp per kWh(.08 USD per kWh * 12000 rp to USD exchange)
# 12000 rp to 1 usd was the exchange rate at time of survey along with reported .08USD per kwh
def Asei_kwh_conversion(cv):
    converted = (cv['PLN_expenditure_monthly']/(960))
    return converted

survey['Asei_kWh'] = survey.apply(Asei_kwh_conversion, axis=1)
```

```python
# calculates the approximate kwh average per household per month
df = survey.groupby(['access_type'])[['PLN_expenditure_monthly','Asei_kWh']].mean()
df.reset_index(inplace=True)
df
```

daily metered energy reflects about 200-250 kwh per day for approximately 40 households which is about 5 kwh per houshold per day or 150 kwh per household per month.

In this second analysis it becomes clear that only households with appliances have reported expenditures which now seems obvious, so maybe there is some correlation between appliance ownership and energy use here.
